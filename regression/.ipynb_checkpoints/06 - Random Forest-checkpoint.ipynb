# Random Forest

Ensemble Learning - quando usa vários algoritimos ou o mesmo, várias vezes, para se obter o resultado.

Passos do Random Forest:

1. Pega K pontos aleatoriamente do training set
2. Monta uma árvore de decisões para esses pontos
3. Escolhe um número _NTREE_ de árvores que queremos montar e repete os passos 1 e 2
4. Para cada novo _data point_, pegar os valores de Y de todos os valores previstos para Y das árvores





```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:,2].values
```

```python
from sklearn.ensemble import RandomForestRegressor

# Cria um modelo com no maximo 10 arvores
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X,y)
```

```python
# Aplicar o modelo no valor 6.5 que queremos prever
y_pred = regressor.predict(6.5)

# Resposta
y_pred
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, regressor.predict(X), color = 'blue')
plt.title('Random Forrest')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

O ponto do CEO é considerado como um _outlier_ pelo SVR, então ele exclui e se concentra nos outros pontos para traçar a curva.

```python
# Curva de predicao mais detalhada
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))

plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Random Forest')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

Como temos mais árvores, temos mais 'degraus' no modelo.

Vamos testar agora com mais árvores.

```python
# Cria um modelo com no maximo 100 arvores
regressor100 = RandomForestRegressor(n_estimators = 100, random_state = 0)
regressor100.fit(X,y)

# Aplicar o modelo no valor 6.5 que queremos prever
y_pred100 = regressor100.predict(6.5)

# Resposta
y_pred100

```

```python
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor100.predict(X_grid), color = 'blue')
plt.title('Random Forest (100 arvores)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Cria um modelo com no maximo 300 arvores
regressor300 = RandomForestRegressor(n_estimators = 300, random_state = 0)
regressor300.fit(X,y)

# Aplicar o modelo no valor 6.5 que queremos prever
y_pred300 = regressor300.predict(6.5)

# Resposta
y_pred300


```

```python
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor300.predict(X_grid), color = 'blue')
plt.title('Random Forest (300 arvores)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```
