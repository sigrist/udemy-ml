## Building a model

Need to decide which one to keep and which one to remove.

1. Garbage in = Garbage out. Some features are garbage and generate wrong results
2. Too many things is hard to explain. Keep the most important are easy to explain

How to select the right features?

1. All-in
2. Backward Elimination
3. Forward Selection
4. Bidirectional Elimination
5. Score Comparison


### All-in 

Put all the variables in the model.

* Priow knowledge - when you have the knowledge of the features
* You have to - when is not your decision or you are obligate to use all features
* Preparing for Backward Elimination

### Backward Elimination

1. Select a significant level to stay in the model. ex (SL = 0.05)
2. Fit the full model with all possible predictors
3. Consider the predictor with the *highest* P-value. If P > SL, go to STEP 4, otherwise go to *FIN*
4. Remove the predictor
5. Fit model without this variable. Then back to step 3 until you get the highest P value

*FIN*: you model is ready


### Forward Selection

1. Select a significant level to stay in the model. ex (SL = 0.05)
2. Fit all simple regression models y ~ xn. Select the one with the lowest P-value
3. Keep this variable and fit all possible models with one extra predictor added to the ones(s) you already have
4. Consider the predictor with the *lowest* P-value. If P < SL, go to Step 3, otherwise go to *FIN*


*FIN*: Keep the previous model

### Birectional Elimination

1. Select the significante level to enter and stay to in the model. SLENTER = 0.05, SLSTAY = 0.05
2. Perform the next step of Forward Selection (new variables must have P < SLENTER to enter)
3. Perform ALL steps of Backward Elimination (old variables must have P < SLSTAY to stay)
4. No new variables can enter and no old variables can exit

*FIN*: Your model is ready

### All Possible Models (Score Comparison)

1. Select a criterion of goodness of fit (ef Akaike criterion)
2. Construct all possible regression models: 2^N - 1 total combinations
3. Select the one with the best criterion

*FIN*: Your model is ready

This is hard to do. A dataset with 10 features will have 1.023 models

## Multiple Linear Regression in Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
dataset = pd.read_csv('50_Startups.csv')
dataset.head()
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 4].values

# Apply the one hot encoder in the State feature
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Apply the label enconder to transform the labels into numbers
labelEncoderX = LabelEncoder()
X[:, 3]= labelEncoderX.fit_transform(X[:, 3])

# Apply the OneHotEncoder to transform the numbers into columns
oneHotEncoder = OneHotEncoder(categorical_features = [3])
X = oneHotEncoder.fit_transform(X).toarray()

# Avoiding the dummy variable trap
X = X[:, 1:] # Remove the first column from X

# Split the dataset into train and test
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
## Fitting Multiple Linear Regression to the Training Set
from sklearn.linear_model import LinearRegression

# Create the model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

### Predicting the values

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
pd.DataFrame({'pred': y_pred, 'real': y_test, 'diff': y_pred - y_test})
```

```python
import statsmodels.formula.api as sm
```

Remember the formula:

y = b0 + b1x1 + b2x2 + ... + bn*xn

There is no x0 for b0, because b0 is a constant. But to use the *sm*, we need to create a x0 with value 1.

```python
# Create an array of ones (arr) and add the X matrix into it
X = np.append(arr = np.ones((50,1)).astype(int), values = X, axis = 1)
```

```python
## Step 2 - Fit the full model wiah all variables

# Original with all features
X_opt = X[:, [0,1,2,3,4,5]]

# Create a new regressor from SM Class
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
```

```python
## Step 3 - Consider the predictor with the highest P-Value

regressor_OLS.summary()
```
