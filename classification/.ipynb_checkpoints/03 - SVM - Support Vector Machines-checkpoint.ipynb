# SVM

Como separar dois grupos de observações já classificadas?

Traçar uma linha entre os dois grupos, mas como achar a melhor linha?


## Porque SVM é tão popular?

Imagine tentando separar maçãs de laranjas.

o SVM encontra a maça mais padrão, ou seja, a mais longe das laranjas, e a mesma coisa para as laranjas, acha a mais longe das maçãs.

Depois, ele procura pelo oposto, as maçãs que mais parecem laranjas e as laranjas que mais parece maçãs.

Esses pontos são os _Support Vectors_, que irão ser a base para traçar a linha entre as categorias, pois elas estão bem perto do limite entre os grupos.



## Python

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create the dataset and X and y datasets
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:,[2,3]].values
y = dataset.iloc[:,4].values
```

```python
# Split the dataset into train and test
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
# Feature Scalling - necessario para o SVR
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
# Fitting SVM to the Training set
from sklearn.svm import SVC

# Por padrao, vamos usar o kernel linear
# Por padrao, a distancia é a manhattan (L1), mas queremos a Euclidiana (L2), portanto, p = 2 e metric = minkowski
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
# Predicting the test set results
y_pred = classifier.predict(X_test)
y_pred
```

### Evaluating the results

Criando a _Confusion Matrix_

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm
```

### Visualizando os resultados

```python
from matplotlib.colors import ListedColormap

X_set, y_set = X_train, y_train

X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() -1, stop = X_set[:, 0].max() +1, step = 0.01), 
np.arange(start = X_set[:, 1].min() -1, stop = X_set[:, 1].max() +1, step = 0.01))
    
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
            alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i,j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j,0], X_set[y_set == j, 1], c = ListedColormap(('yellow', 'blue'))(i), label = j)
    
plt.title('SVM Linear (Training Set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

* Os pontos são as pontos de observação do training set
* A região vermelha é a prevista de purchased 0 (não comprou) e a verde é 1 (que comprou)
* Pontos amarelos não compraram, e os azuis compraram

Observações:

* Os que tem pouca idade, e pouco salário, não compraram o carro
* Os mais velhos, e com salário maior, compraram o carro


Agora para os dados de teste:

```python
from matplotlib.colors import ListedColormap

X_set, y_set = X_test, y_test

# Pega o menor do valor do X_Set e do maior valor, e incrementa 0.01
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() -1, stop = X_set[:, 0].max() +1, step = 0.01), 
np.arange(start = X_set[:, 1].min() -1, stop = X_set[:, 1].max() +1, step = 0.01))

# Para cada pixel do grafico, aplica o predict, para pintar de vermelho ou verde
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
            alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i,j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j,0], X_set[y_set == j, 1], c = ListedColormap(('yellow', 'blue'))(i), label = j)
    
plt.title('SVM Linear (Test Set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
