# K-Means

Como o _K-Means_ funciona?

1. Escolha o número de _K_ grupos
2. Selecione aleatóriamoente _K_ pontos, que serão os _centroids_ dos grupos (esses pontos não necessariamente fazem parte do dataset)
3. Atribua para cada data point ao centroid mais perto utilizando algum algoritmo de cálculo de distancia (Euclidian, Manhattan, etc)
4. Calcule e coloque o novo centroid para cada cluster (grupo)
5. reatribua para cada data point o novo centroid mais perto. Se algum item se alterou, volte para o passo 4, senão o algoritmo terminou

## Escolhendo o número correto de cluster (K)

Imagine o mesmo grupo de observações abaixo:

![](kmeans01.png)

Como saber qual é o número correto de cluster? No caso, pode ser 2 ou 3.

Temos algumas métricas para avaliar. 


$$ WCSS = \sum\limits_{Pi in Cluster 1} distance(Pi, C1)^2 + \sum\limits_{Pi in Cluster 2} distance(Pi, C2)^2 + \sum\limits_{Pi in Cluster 3} distance(Pi, C3)^2$$


Pegar todos os pontos do cluster e acumule a soma da distancia de todos os pontos do cluster com o centroid e eleve o resultado ao quadrado.

Utilizando o exemplo acima. Se utilizarmos `K = 1`, iremos ter o seguinte resultado:

Para `K = 2`, teremos o seguinte:

Para `K = 3`, teremos o seguinte:

![](kmeans07.png)

Abaixo é o resultado dos resultados de `WCSS`. Para achar o melhor número de cluster, utilizamos p _Elbow Method_ (cotovelo), que é o ponto onde temos a maior alteração da tendência do valow WCSS:

## Python


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Import the dataset
dataset = pd.read_csv('Mall_Customers.csv')
dataset.head()

##
# O dataset é baseado pelo genero, idade e salário anual e um score de quanto o cliente gasta
# Não sabemos quantos grupos de clientes temos
```

```python
## Vamos usar o Annual Income e o Spending Score 

# Criar uma matriz com somente esses dados
X = dataset.iloc[:,[3,4]].values
```

```python
## Como não sabemos a quantidade de cluster, vamos utilizar o metodo elbow
from sklearn.cluster import KMeans

wcss = []

# Vamos tentar de 1 ate 10 cluster (11 pq o range nao inclui o ultimo item)
for i in range(1,11):
    # Criar um cluster de i grupos, utilizando a inicializacao K-Means++ (evitar random trap), 
    # maximo de 300 iteracoes e n_init = 10 é o maximo de vezes que o algortimo vai rodar as inicializacoes 
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    
    # aplicar nosso dataset no cluster
    kmeans.fit(X)
    
    # Agora calcular o WCSS
    # A biblioteca sklearn ja faz esse calculo, e é chamdo de inertia_
    wcss.append(kmeans.inertia_)
    
    
# Agora vamos plotar os resultados

plt.plot(range(1,11), wcss)
plt.title('The Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
    
    
```

Vemos que o cotovelo está em `K = 5`

```python
# aplicando o K-Means com K = 5

kmeans = KMeans(n_clusters = 5, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)

# Vamos agora usar o metodo fit_predict, que ele retornara o resultado dos agrupamentos
y_kmeans = kmeans.fit_predict(X)
y_kmeans
```

```python
# Visualizando os resultados

# Pontos, de acordo com o grupo
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'yellow', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')

# agora os centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'green', label = 'Centroids')
plt.title('Cluster of clients')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

Avaliando os resultados, podemos agora dar nomes mais adequados para os clusters:

Cluster1: Tem um ganho anual grande, mas não gastam tanto. São cuidadosos. Label: **Careful**
Cluster2: Ganham na média e gastam na média. Label: **Standard**
Cluster3: Ganham bem e gastam bem. Label: **Target**
Cluster4: Ganham pouco, mas gastam bem. Label: **Careless**
Cluster5: Ganham pouco, gastam pouco. Label: **Sensible**
